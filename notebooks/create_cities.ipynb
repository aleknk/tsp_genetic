```python
import numpy as np
import matplotlib.pyplot as plt
from pytsp.cities import generate_cities, plot_cities
```

```python
# Set params
save_path = "../data/cities_example.npy"
n, min_val, max_val = 50, 0, 100

# Create cities
cities = generate_cities(n=n, min_val=min_val, max_val=max_val)

# Plot cities
fig, ax = plt.subplots(figsize=(6,6))
plot_cities(cities, ax=ax)

# Save cities
np.save(save_path, cities)
```
